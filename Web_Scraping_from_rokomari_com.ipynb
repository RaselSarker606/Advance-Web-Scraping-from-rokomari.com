IMPORTING SOME LIBRARIES

```python
import requests
from bs4 import BeautifulSoup
```

Now Scraping from rokomari.com

```python
list = []
for i in range(1,20):
  url = f'https://www.rokomari.com/book/category/410/science-fiction?ref=h_cl16&page={i}'
  r = requests.get(url)

  formattedcontent = BeautifulSoup(r.content,'html5lib')

  table=formattedcontent.findAll('div',attrs={'class':'book-text-area'})

  for i in table:
    title = i.find('h4').text
    author = i.find('p').text
    data = {'title' : title,
            'author' : author}

    list.append(data)

    print(data)

```

```python
print(len(list))
```

```python
import pandas as pd
df = pd.DataFrame(list)
df

```
